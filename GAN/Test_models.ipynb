```python
import torch
from torch import autograd
import matplotlib.pyplot as plt
from torch.autograd import Variable
from models.pulse2pulse import WaveGANGenerator as Pulse2PuseGenerator
```

```python
path = "output/test_exp_1/checkpoints/_epoch4000.pt"
netG = Pulse2PuseGenerator(model_size=50, ngpus=1,num_channels=2,upsample=True)
chkpnt = torch.load(path, map_location="cpu")
netG.load_state_dict(chkpnt["netG_state_dict"])
```

```python
num_data = 10
frame_size = 512
ch = 0

for i in range(num_data):
    noise = torch.Tensor(1, 2, 5000).uniform_(-1, 1)
    noise_Var = Variable(noise, requires_grad=False)
    fake = autograd.Variable(netG(noise_Var).data)
    plt.plot(fake[0][ch][0:frame_size])
    plt.show()
```
